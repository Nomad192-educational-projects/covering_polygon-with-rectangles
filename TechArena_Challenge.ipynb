```python
from matplotlib.patches import Polygon
import matplotlib.pyplot as plt
```

```python
def drawPoly(ax, coord):
  polygon = Polygon(coord, closed=True, fill=None)
  ax.add_patch(polygon)
  ax.autoscale()
```

```python
def getCoord(test):
  coord = []
  n = 0.0
  c1 = 0.0
  c2 = 0.0 
  with open("0" + str(test) + ".txt", "r") as f:
    n, c1, c2 = f.readline().split()
    for line in f.readlines():
      cur_x, cur_y = line.split()
      coord.append((cur_x, cur_y))
  return n, c1, c2, coord
```

```python
def showTest(test):
  fig,(ax) = plt.subplots(figsize=(8,8), ncols=1)

  n, c1, c2, coord = getCoord(test)
  drawPoly(ax, coord)

  fig.show()
```

```python
showTest(1)
```

```python
def getCoord_file(test_file):
  coord = []
  n = 0.0
  c1 = 0.0
  c2 = 0.0 
  with open(test_file, "r") as f:
    n, c1, c2 = f.readline().split()
    for line in f.readlines():
      cur_x, cur_y = line.split()
      coord.append((cur_x, cur_y))
  return n, c1, c2, coord
```

```python
from matplotlib.legend_handler import Rectangle
def showTest_withAnswer_with_file(test_file, answer_file):
  fig,(ax) = plt.subplots(figsize=(16,16), ncols=1)

  n, c1, c2, coord = getCoord_file(test_file)
  drawPoly(ax, coord)

  with open(answer_file, "r") as f:
    n = f.readline().split()
    for line in f.readlines():
      l_x, b_y, r_x, t_y = line.split()
      drawRec(ax, float(l_x), float(b_y), float(r_x), float(t_y))

  drawPoly(ax, coord) 

  fig.show()
```
